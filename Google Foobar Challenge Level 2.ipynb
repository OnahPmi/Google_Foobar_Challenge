```python
# This is the solution that was submitted

from fractions import Fraction  
def solution(pegs):
    """
    This function implements the following algorithm
    
    Given a list containing peg positions sorted in ascending order
    
    pegs = [pegs[0], pegs[1], pegs[2], ..., pegs[n]]
    
    After researching for a while, the following relationship between gear radius and peg position was found
    
    first_gear_radius = 2/3*((-pegs[0]-pegs[n]) + 2*(pegs[1]-pegs[2]+...+pegs[n-2]-pegs[n-1])) if n is even
    first_gear_radius = 2*((-pegs[0]+pegs[n]) + 2*(pegs[1]-pegs[2]+pegs[3]-...-pegs[n-2]+pegs[n-1])) if n is odd
    
    next_gear_radius = successive_pegs_distance - current_gear_radius
    Where, successive_pegs_distance = next_peg_position - current_peg_position
    
    E.g second_gear_radius = (pegs[1] - pegs[0]) - first_gear_radius
    
    n = number of peg positions in pegs list
    
    ------Conditions to consider-----------
    
    1. n must be between 2 and 20,
    2. peg_position must be between 1 and 10000,
    3. first_gear_radius = 2*last_gear_radius. Thus, first_gear_radius must be greater than or equal to 2
    4. All gear radii must be greater than 1.
    """
    n = len(pegs)
    
    # Checks if the pegs list length is within allowed limit (1st condition)
    if n < 2 or n > 20:
        return [-1, -1]
        
    # Check for invalid peg_positions in the pegs list (2nd condition)
    if len([peg_position for peg_position in pegs if peg_position > 10000 or peg_position < 1]) != 0:
        return [-1, -1]
        
       # If no violation of the conditions above, then compute the first_gear_radius radii as follows:
    filterd_pegs = [peg_position for peg_position in pegs if peg_position != pegs[0] and peg_position != pegs[-1]]
    first_gear_radius_term = (-pegs[0]+pegs[-1]) if (n % 2 == 0) else (-pegs[0]-pegs[-1])
    for peg_position in filterd_pegs:
        first_gear_radius_term += float((2*peg_position) if filterd_pegs.index(peg_position) % 2 == 0 else (2*-peg_position))
    first_gear_radius = 2*first_gear_radius_term/3 if (n % 2 == 0) else 2*first_gear_radius_term
    
       # Express the radius as a fraction in in its simplest form
    first_gear_radius = Fraction(first_gear_radius).limit_denominator()
     
    # Check each gear radius to make sure none is < 1 (3rd condition)
    current_gear_radius = first_gear_radius
    index = 0 # initialize index
    while index < n-1:
        successive_pegs_distance = pegs[index+1] - pegs[index]
        next_gear_radius = successive_pegs_distance - current_gear_radius
        if (next_gear_radius < 1):
            return [-1,-1]
            index+=1
        else:
            current_gear_radius = next_gear_radius
            index+=1
            
    # Check if first_gear_radius < 2 (4th condition) 
    if first_gear_radius < 2:
        return [-1, -1]
    else:
        return [first_gear_radius.numerator, first_gear_radius.denominator]
    
test_cases()
```

```python
# Method 3

from fractions import Fraction
def solution(pegs):

    n = len(pegs)
        
    filterd_pegs = [num for num in pegs if num != pegs[0] and num != pegs[-1]]
    middle_term = 0
    for num in filterd_pegs:
        middle_term += num if filterd_pegs.index(num) % 2 == 0 else - num  
    if n % 2 == 0:
        term = -pegs[0]+pegs[-1]+ 2* middle_term
    else:
        term = -pegs[0]-pegs[-1]+ 2* middle_term
    
    r1 = Fraction(2 * (float(term)/3 if (n % 2 == 0) else term)).limit_denominator()
    
    # now that we have the radius of the first gear, we should again check the input array of pegs to verify that
    # the pegs radius' is atleast 1.
    # since for valid results, LastGearRadius >= 1 and FirstGearRadius = 2 * LastGearRadius
    # thus for valid results FirstGearRadius >= 2

    if r1 < 2:
        return [-1,-1]

    currentRadius = r1
    for index in range(0, n-2):
        CenterDistance = pegs[index+1] - pegs[index]
        NextRadius = CenterDistance - currentRadius
        if (currentRadius < 1 or NextRadius < 1):
            return [-1,-1]
        else:
            currentRadius = NextRadius
  
    a_b = Fraction(r1).limit_denominator()
    a, b = a_b.numerator, a_b.denominator

    return [a, b]

test_cases()
```
